import packages

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

load sample image

```python
pic = cv2.imread('together.jpg')
```

display loaded image

```python
plt.title('Orignal')
plt.imshow(pic)
```

crop the image

```python

crop_image = pic[0:175, 0:250]

```

display the cropped image

```python
plt.title('Crop')
plt.imshow(crop_image)
```

load another sample image

```python
pic2 = cv2.imread('nature.jpg')
```

display another loaded image

```python
plt.title('Orignal')
plt.imshow(pic2)
```

crop 2nd sample image

```python
crop2 = pic2[0:175, 0:250]
```

```python
plt.title('cropped')
plt.imshow(crop2)
```
